```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def listToTable(lis):
    table = []
    a = np.asarray(lis)
    for j in range(0, len(lis)):
        a1 = a[j][0:-2].split(' ')
        a2 = map(float, a1)
        a3 = np.asarray(a2)
        table.append(a3)
    table = np.asarray(table)
    return table
def linear(t, k, b):
    return k*t + b
def getFit(t, totDisplacement2):
    popt, pcov = curve_fit(linear, t, totDisplacement2, absolute_sigma=True)
    k, b = popt
    sigma_k, sigma_b = np.sqrt(np.diag(pcov))
    return k, sigma_k, b, sigma_b
def plotFit(t, totDisplacement2, color, label):
    plt.plot(t, totDisplacement2, color=color, marker='o', linewidth=0, label=label)
    plt.plot(t, linear(t, getFit(t, totDisplacement2)[0], getFit(t, totDisplacement2)[2]), color=color, linewidth=1)
    return getFit(t, totDisplacement2)[0]
def getMeanD(dr2_array, tau, dimensions=2):
    D = np.mean(dr2_array)/(2*dimensions*tau)
    return D
def getTheoryD(eta, d):
    kB = 1.38e-23
    T = 293
    D = kB * T / (3 * np.pi * eta * d)
    return D
def getTableVar(table, var):
    if var == 'x':
        return table[:,0]
    elif var == 'y':
        return table[:,1]
    elif var == 't':
        return table[:,2]
    elif var == 'dx':
        return table[:,3]
    elif var == 'dy':
        return table[:,4]
    elif var == 'dt':
        return table[:,5]
    elif var == 'dr2':
        return table[:,6]
    elif var == 'totDisplacement2':
        return table[:,7]
    elif var == 'tau':
        return table[:,5][1]
    else:
        return False
```
