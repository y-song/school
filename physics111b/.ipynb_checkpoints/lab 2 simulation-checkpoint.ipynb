```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
displacement = np.random.normal(size=N)
plt.plot(displacement)
```

```python
plt.hist(displacement, bins=25)
```

```python
x = np.cumsum(displacement)
plt.plot(x);
plt.ylabel('position')
plt.xlabel('time step')
plt.title('Position of 1D Particle versus Time')
```

```python
x = np.cumsum(np.random.normal(size=N))
y = np.cumsum(np.random.normal(size=N))
plt.plot(x, y)
plt.ylabel('Y Position')
plt.xlabel('X Position')
plt.title('position versus time in 2D')
```

```python
dsquared = x**2 + y**2
plt.plot(dsquared)
```

```python
d    = 6.1e-6
eta  = np.array([1.66e-3, 2.50e-3, 4.65e-3, 13.2e-3])
kB   = 1.38e-23
T    = 293
D    = kB * T / (3 * np.pi * eta * d)
```

```python
dimensions = 2
tau = .1
time = tau * np.arange(N)
for i in range(0,4):
    k = np.sqrt(D[i] * dimensions * tau)
    dx = k * np.random.normal(size=N)
    dy = k * np.random.normal(size=N)
    x = np.cumsum(dx)
    y = np.cumsum(dy)
    dSquaredDisplacement = dx ** 2 + dy ** 2
    squaredDisplacement = x ** 2 + y ** 2
    plt.plot(x,y)
plt.title('Particle Track of a Single Simulated Particle')
```

```python
color_array = ['r','g','b','purple']
label_array = ['$\eta=1.7cP$','$\eta=2.5cP$','$\eta=4.7cP$','$\eta=13cP$']
```

```python
fig = plt.figure(figsize=(8,6))
dimensions = 2
tau = .1
time = tau * np.arange(N)
for i in range(0,4):
    k = np.sqrt(D[i] * dimensions * tau)
    dx = k * np.random.normal(size=N)
    dy = k * np.random.normal(size=N)
    x = np.cumsum(dx)
    y = np.cumsum(dy)
    dSquaredDisplacement = dx ** 2 + dy ** 2
    squaredDisplacement = x ** 2 + y ** 2
    plt.plot(time, np.arange(0,N)*2*k**2, color=color_array[i],label=label_array[i])
    plt.plot(time, squaredDisplacement, color=color_array[i])
plt.xlabel('Time')
plt.ylabel('Displacement Squared')
plt.title('Displacement Squared versus Time for 1 Particle in 2 Dimensions')
plt.legend()
```
